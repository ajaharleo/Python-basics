```python
# 4: dump all the data into car table
```

```python
import csv
```

```python
 with open('car.data','r') as f:
        car_data = csv.reader(f,delimiter = '\n')
        for i in car_data:
            rows = i[0].split(',')
            cursor.execute(f'insert into cardataset.car values (\"{rows[0]}\",\"{rows[1]}\",\"{(rows[2])}\",\"{(rows[3])}\",\"{rows[4]}\",\"{rows[5]}\",\"{rows[6]}\")')
mydb.commit()
```
