# Histogram

```
figure,axes=plt.subplots(2,2,figsize=(20,10))
axes[0,0].set_title('Sepal length')
axes[0,0].hist(iris['sepal_length'],bins=10)

axes[0,1].set_title('Sepal width')
axes[0,1].hist(iris['sepal_width'],bins=10)

axes[1,0].set_title('Petal length')
axes[1,0].hist(iris['petal_length'],bins=10)

axes[1,1].set_title('Petal width')
axes[1,1].hist(iris['petal_width'],bins=10)
```

The highest frequecy of Sepal length is around 27 between 5.5 to 6.0

The highest frequency of Sepal width is more than 35 between 3.0 to 3.3

The highest frequency of Petal length is more than 35 between 1 to 1.5

The highest frequency of Petal width is 40 between 0.0 to 0.3 




```
sns.FacetGrid(iris,hue='species',height=6).map(sns.distplot,'sepal_length').add_legend()
sns.FacetGrid(iris,hue='species',height=6).map(sns.distplot,'sepal_width').add_legend()
sns.FacetGrid(iris,hue='species',height=6).map(sns.distplot,'petal_length').add_legend()
sns.FacetGrid(iris,hue='species',height=6).map(sns.distplot,'petal_width').add_legend()
```

We can clearly see that our Setosa flowers are easily separable from Versicolor and virginca.

```
sns.pairplot(iris,hue='species',height=4)
```

We can't differentiate our species with Sepal length and Sepal width but Petal length and Petal width are the most important features of these four.We can easily differentiate Setosa flowers from others as it has minimum values of Petal length and Petal width.Except some mix ups in Versicolor and Verginca, we can separate these two with Petal length and Petal width.

```
iris.corr()
```

We see that Sepal length have good correlation with Petal length and Petal width. Petal length have highest correlation with Petal width.
